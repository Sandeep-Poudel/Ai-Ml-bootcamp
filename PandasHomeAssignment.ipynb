## Loc vs iLoc
> - Loc ma column ko thau man column name list ma diney ho, index dina paudaina.
> - Iloc ma coulumn ko index diney ho name dina puadaina
> - While slicing, For example [25:30]
>> - 30 is included in loc.
>> - 30 is not included in Iloc.


```python
df.loc[25:30, ['sex', 'time']]
```

```python
df.iloc[25:30, 2:6:3]
```
