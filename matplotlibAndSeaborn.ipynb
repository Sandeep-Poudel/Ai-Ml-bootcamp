```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

## Line plot

```python
month = ['Jan', 'Feb', 'Mar', "Apr", 'May']
growth = [3.5, 5, 5.3, 5, 6]
```

```python
plt.plot(month,growth)
```

#### Connect with line

```python
plt.plot(month, growth, "o-")
```

#### Changing the color of the graph

```python
plt.plot(month, growth, "o-", color = "green")

#Give label to the graph
plt.xlabel("Month")
plt.ylabel("Growth in percentage")

#to remove the text
plt.show()
```

```python
plt.plot(month, growth, "o-", color = "blue")

#Give label to the graph
plt.xlabel("Month")
plt.ylabel("Growth in percentage")

#give the title to the plot
plt.title("Growth by Month")

#to remove the text
plt.show()
```

## BarPlot
- ### For catagorical data

```python
## Loading dataset from seaborn
df = sns.load_dataset('iris')
df.head()
```

```python
df.species.unique()
```

```python
df.species.value_counts()
```

```python
s = df.species.value_counts()
s
```

```python
plt.bar(x = s.index, height = s.values, color=['red','blue', 'green'])
plt.xlabel("Species")
plt.ylabel("Frequency/count")
plt.show()
```

#### The above data can be represented using seaborn in the following way

```python
## Barplot of a column
sns.countplot(x = "species",data = df)
```

## Pie Chart

```python
s = df.species.value_counts()
plt.pie(labels = s.index , x = s.values, autopct="%.2f%%")     #label = category
plt.show()
```

## Scatterplot
- Shows relationship between two variable

```python
x = df['petal_length']
y = df['petal_width']
plt.scatter(x=x, y=y)
```

#### Using Seaborn

```python
sns.scatterplot(x = df['petal_length'] , y = df['petal_width'], hue= df['species'])
```

#### To calculate pearson corelation coefficient 

```python
from scipy.stats import pearsonr 
r = pearsonr(df['petal_length'], df['petal_width'])
r.statistic
```

#### Pandas has a corr() method that does the above work for all the datas

```python
df.corr(numeric_only = True)
```

```python
sns.heatmap(df.corr(numeric_only=True))
```

```python
# To print the values in the heatmap
sns.heatmap(df.corr(numeric_only=True), annot=True)
```
